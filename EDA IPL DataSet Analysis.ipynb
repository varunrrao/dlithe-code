```python
#Team with most number of wins
matches = pd.read_csv('matches.csv')
data = matches.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
#Mumbai Indians have the highest number of wins across all the seasons
```

```python
#Who mon most Top player of the match awards
top_players = matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
sns.barplot(x = top_players.index,y=top_players,orient='v'); #palette="Blues");
plt.show()
#Chris Gayle won the most number of player of the match awards
```

```python
#How many Toss winning teams have won the matches?
ss = matches['toss_winner']==matches['winner']
ss.groupby(ss).size()
round(ss.groupby(ss).size()/ss.count()*100,2)
sns.countplot(ss);
#Probability of winning the match after winning the toss is 51.5%
```

```python
#No of Tosses won by each team
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
match_data=pd.read_csv("IPL Matches 2008-2020.csv") 
toss=match_data['toss_winner'].value_counts()
ax=plt.axes()
ax.set(facecolor="grey")
sns.set(rc={'figure.figsize':(15,10)},style='darkgrid')
ax.set_title('No of tosses won by each team',fontsize=15,fontweight="bold")
sns.barplot(y=toss.index,x=toss,orient='h',palette="icefire",saturation=1)
plt.xlabel('# of Tosses won')
plt.ylabel('Teams')
plt.show()
#Mumbai Indians won the highest number of tosses
```

```python
plt.figure(figsize=(12,4))
sns.countplot(match_data.toss_decision[match_data.toss_winner==match_data.winner])
plt.show()
#Choosing to field first after winning the toss has a higher winning probability
```

```python
#Top 5 Man of the match Awards
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
temp_data=match['player_of_match'].value_counts().head()
print(temp_data)
sns.barplot(x=temp_data.index,y=temp_data.values,data=match)
plt.title("Top 5 MoM")
plt.xticks(rotation=90)
plt.xlabel("Match Count")
plt.ylabel("Player")
plt.show()
```

```python
#Top 5 Batsman in the world
top_batsman=delivery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values('batsman_runs', ascending=False).head(10)
top_batsman.set_index('batsman', inplace=True)
top_batsman.plot(kind='bar')
#Raina,Kohli,R Sharma,G gambhir and D warner are currently the 5 best batsman in IPL
```

```python
#Bowler Piyush Chawla against other team wise Performance
mask=delivery['bowler']=='PP Chawla'
delivery[mask].groupby('batting_team')['total_runs'].agg('sum').plot(kind='bar')
#he should not bowl against MI,RCB,DD AND CSK
```

```python
#Heatmap of Over-wise Batting Performance of Each Team in the IPL (2008–2020)
delivery6=delivery[mask]
delivery6=delivery6[['batting_team','over','batsman_runs']]
x=delivery6.pivot_table(values='batsman_runs', index='batting_team', columns='over', aggfunc='count')
sns.heatmap(x, cmap='summer')
#Mumbai Indians have played well in the middle and death overs but not so well in the powerplsy
```

```python
#Created a new attribute called Season in Matches 2008- 2020 data frame
match_data['season']=pd.DatetimeIndex(match_data['date']).year
match_data.head()
```

```python
#Matches Played in each season
match_per_season=match_data.groupby(['season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(match_data['season'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('season',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Total matches played in each season',fontsize=10,fontweight="bold")
#Most matches were played on 2013
```

```python
season_data=match_data[['id','season']].merge(ball_data,left_on='id',right_on='id',how='left').drop('id',axis=1)
season_data.head()
```

```python
#Plotting a line graph to display most runs scored in each season
season=season_data.groupby(['season'])['total_runs'].sum().reset_index()
p=season.set_index('season')
ax=plt.axes()
ax.set(facecolor="grey")
sns.lineplot(data=p,palette="magma")
plt.title('Total runs in each season',fontsize=12,fontweight="bold")
plt.show()
#Most runs were scored in 2013 season
```
